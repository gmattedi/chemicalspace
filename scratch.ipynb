```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from chemicalspace import ChemicalSpace
from IPython.display import display
from rdkit import Chem
from rdkit.Chem import Draw

sns.set_style("darkgrid")

Draw.rdDepictor.SetPreferCoordGen(True)
```

```python
space = ChemicalSpace.from_smi("data/inputs1.smi")
space
```

```python
proj = space.project()
```

```python
label = space.cluster(method="scaffold")
```

```python
len(set(label))
```

```python
for i in sorted(set(label)):
    mask = label == i
    plt.scatter(*proj[mask].T, edgecolor="k")
plt.show()
```
